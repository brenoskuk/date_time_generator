## This program generates random entry and leaving times during a span of several days

### First asks parameters for the user

```python
import datetime
```

```python
date_entry = input('Enter a start date in DD/MM/YYYY format: ')
day, month, year = map(int, date_entry.split('/'))
start_date = datetime.date(year, month, day)
date_entry = input('Enter a end date in DD/MM/YYYY format: ')
day, month, year = map(int, date_entry.split('/'))
end_date = datetime.date(year, month, day)
```

```python
enter_hour = int(input('Enter a start hour (eg.: 9am is 9): '))
enter_minute = int(input('Enter a start minutes (eg.: 30 min is 30): '))
leave_hour = int(input('Enter leave hour (eg.: 8pm is 20): '))
leave_minute = int(input('Enter leave minutes: '))
```

```python
lunch_start_hour = int(input('Enter lunch start hour: '))
lunch_start_minute = int(input('Enter lunch start minute: '))
lunch_time = int(input('Enter lunch time (can be set to zero)(eg.: 1 hour of lunchtime is 1): '))
```

```python
good_variation = int(input('Enter the good variation in minutes (eg.: If set to 5 and you are supposed to arrive 9am you could arrive be early at 8:55): ')) 
bad_variation = int(input('Enter the bad variation in minutes (eg.: If set to 5 and you are supposed to arrive 9am you could arrive late at 9:05): ')) 
```

```python
date = datetime.datetime.strptime('1 Jan 2012', '%d %b %Y')
default_enter_time = date.replace(hour=enter_hour, minute=enter_minute)
default_lunch_start_time = date.replace(hour=lunch_start_hour, minute=lunch_start_minute)
default_lunch_return_time = date.replace(hour=lunch_start_hour+lunch_time, minute=enter_minute)
default_leave_time = date.replace(hour=leave_hour, minute=leave_minute)
```

### Creating a list of holydais manually (in his case for the city of s√£o paulo in 2019 from february to may

```python
#in order: carnaval (4/3/2019 to 6/3/2019), sexta feira santa (19/4/2019), dia do trabalho (01/05/2019)
holidays = [datetime.date(2019, 3,4), datetime.date(2019, 3,5), datetime.date(2019, 3,6), datetime.date(2019, 4,19),datetime.date(2019, 5,1),]

print (holidays) # Will print "True" or "False"
```

### Then generates random entry times, leave times, lunch times for given parameters

```python
import random
delta = random.randint(-good_variation,bad_variation)
```

```python
current = start_date
while current <= end_date:
    
    if current in holidays:
        print ('holiday\n ')
        
    elif current.weekday() < 5:
    
        delta = random.randint(-good_variation,bad_variation)
        enter_time = default_enter_time + datetime.timedelta(minutes = delta)
        delta = random.randint(-bad_variation,good_variation)
        lunch_start_time = default_lunch_start_time + datetime.timedelta(minutes = delta)
        delta = random.randint(-good_variation,bad_variation)
        lunch_return_time = default_lunch_return_time + datetime.timedelta(minutes = delta)
        delta = random.randint(-bad_variation,good_variation)
        leave_time = default_leave_time + datetime.timedelta(minutes = delta)

        print('day: ', current)
        print('entry time: ', enter_time.time())
        print('lunch start time: ', lunch_start_time.time())
        print('lunch return time: ', lunch_return_time.time())
        print('leave time: ', leave_time.time())
        print('\n')
    
    else:
        print ('weekend\n')
    current += datetime.timedelta(days=1)
    
```
